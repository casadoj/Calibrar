```python
import spotpy
import numpy as np
import pandas as pd
import scipy as sp
from netCDF4 import Dataset
import pandas as pd
import os
import sys
from datetime import datetime
import matplotlib.pyplot as plt
import time
import tqdm
```

```python
def hydrological_model(precipitation, q, b):
    Q=precipitation+q+precipitation/b
    return Q
```

```python
precipitation=np.random.randint(600, size=365)
```

```python
caudal_obs=hydrological_model(precipitation,q=8.8787878,b=2.78555)
```

```python
class vic_model(object):
    def __init__(self, precipitation):
        self.precipitation=precipitation
        return
    def run_vic(self,precipitation,q, b ):
        Q=precipitation+q+precipitation/b
        return Q
        
    
class spotpy_setup(object):
    def __init__(self,precipitation,obs,OF):
        self.precipitation=precipitation
        self.vicmodel=vic_model(self.precipitation)
        self.params = [spotpy.parameter.Uniform('q',0,200),
                      spotpy.parameter.Uniform('b',0,200)]
        self.obs=obs
        self.OF=OF
        return
    def parameters(self):
        return spotpy.parameter.generate(self.params)
    def simulation(self,vector):
        simulations= self.vicmodel.run_vic(self.precipitation,vector[0],vector[1])
        return simulations
    def evaluation(self):
        observations=self.obs
        return observations
    def objectivefunction(self,simulation,evaluation):
        OF=self.OF
        if OF == 'nse':
            objFunc = -spotpy.objectivefunctions.nashsutcliffe(evaluation, simulation)
        elif OF == 'bias':
            objFunc =abs(spotpy.objectivefunctions.pbias(evaluation, simulation))
        elif OF == 'rmse':
            objFunc = spotpy.objectivefunctions.rmse(evaluation=evaluation, simulation=simulation)
        elif OF == 'kge':
            objFunc = spotpy.objectivefunctions.kge(evaluation, simulation)
        else:
            print('ERROR. Wrong objective criteria')
        
        return objFunc
def findBestSim(dbPath, OF):
    # import results
    csv = pd.read_csv(dbPath + '.csv')
    results = np.array(csv)

    # get the index (i.e., iteration) with the lowest error
    likes = np.array(csv.like1)
    if OF in ['nse', 'kge']:
        idx = likes.argmax()
    elif OF in ['rmse', 'bias']:
        idx = likes.argmin()  
    else:
        print('ERROR. Wrong objective function')

    # extract the model parameter values for that iteration
    q = results[idx,1]
    b = results[idx,2]
    params = [q, b]
    print(q,b)
    return params
```

```python
results=[]
rep=10000
OF='nse'
sampler=spotpy.algorithms.sceua(spotpy_setup(precipitation,caudal_obs,OF),    dbname='ROPE', dbformat='csv')
sampler.sample(rep,ngs=10)
results.append(sampler.getdata())
params=findBestSim('ROPE',OF)
```
