__Introducción__<br>
Se analiza cómo podría ser posible calibrar la forma en que TETIS simula la evapotranspiración para que ésta reproduzca las observaciones de MODIS.

En la evapotranspiración intervienen dos tanques: la evaporación ($E_6$) desde el tanque T6-interceptación y la evapotranspiración ($ET_1$) desde el tanque T1-estático.
$$ET=E_6+ET_1<=ETP$$

1.  La <font color='steelblue'>__evaporación desde el agua interceptada__</font>:
$$E_6=min[ETP\cdot\lambda_v(t)\cdot FC_2,H_6]$$
Donde $ETP$ es la evapotranspiración potencial definida en el archivo de entrada, $\lambda_v$ es el factor de cultivo mensual para cada tipo de cobertura del suelo y $FC_2$ es el factor corrector de la evapotranspiración (el parámetro calibrable).

2.  La <font color='steelblue'>__evapotranspiración desde el tanque estático__</font>:
$$ET_1=min[(ETP\cdot FC_2-E_6)\cdot\lambda_v(t)\cdot f(\theta),H_1]$$
Donde $f(\theta)$ es la función de humedad que representa el contenido de humedad en el suelo en función de tres umbrales (punto de marchitez, humedad óptima y saturación). Puesto que $ET_1$ depende de la cantidad de agua en el tanque estático ($H_1$), está ligada indirectamente al parámetro $p_1$, que define la infiltración de la precipitación neta en este tanque:
$$I_1=min\left[(X_6+Y_0)\cdot \left(1-\frac{H_1}{H_u\cdot FC_1}\right)^{p_1},H_u\cdot FC_1-H_1\right]$$

Según esta formulación, se puede alterar la evapotranspiración simulada de tres maneras: modificando los factores de cultivo, calibrando $FC_2$ y $p_1$, o en menor medida, modificando la humedad óptima. 

El método más obvio, y el utilizado hasta ahora, es <font color='steelblue'>calibrar el factor corrector $FC_2$</font>, lo cual tiene ciertas limitaciones. La primera es que se calibra la ET para reducir el sesgo del caudal, no para reproducir ET observada. Además, modificando $FC_2$ se puede ajustar la ET de la cuenca en términos globales para ajustar el balance hídrico a escalas temporales grandes, pero no se puede ajustar la variabilidad espacial ni intra-anual causada por el distinto comportamiento de las coberturas vegetales (ciclo vegetativo anual y comportamiento frente a la sequía). Por ello cobra gran importancia el ciclo vegetativo anual definido a través de los factores de cultivo mensuales de cada tipo de cobertura.

Los <font color='steelblue'>factores de cultivo ($\lambda_v(t)$)</font> corresponden a un mapa fijo del ciclo anual de la vegetación. Desde el punto de vista estrictamente teórico, es una aproximación muy simplista, puesto que la vegetación no responde al clima de cada año, sino que la fenología es estática. Más allá de ser una simplificación, la dificultad de definir a priori las curvas de factor de cultivo exigiría que dichas curvas fueran modulables durante la calibración, de manera que al menos se pueda ajustar dicha imagen fija en el año para que la simulación reproduzca la observación. La comparativa de la simulación en TETIS con la observación de MODIS para la cuenca del Deva deja claro que, más allá de haber un sesgo positivo en la simulación de la ET, hay un desfase temporal del ciclo vegetativo; la fase anual de ET en la simulación ocurre con antelación de 1 a 2 meses con respecto a la observación.

Lo ideal sería simular la vegetación de forma dinámica, de manera que el clima de las semanas previas defina el funcionamiento de la vegetación en un día dado. Es decir, que la fenología de la vegetación (su factor de cultivo) cambie año a año en función del clima de dicho año. Puesto que esto no es el objetivo por ahora, la idea sería ser capaces de modificar las curvas de cultivo estáticas para evitar el desfase entre simulación y observación anteriormente comentado. El objetivo de este _notebook_ es idear cómo modular dichas curvas.

Una forma de idear cómo han de ser las curvas del factor de cultivo es a partir de los productos de LAI y cobertura del suelo de MODIS. Se puede extraer un _proxy_ de la curva de cultivo para cada tipo de cobertura. La magnitud absoluta deja de ser importante, puesto que será lo que se calibre con $FC_2$, siendo la clave el valor relativo entre tipos de cobertura.

Utilizar la <font color='steelblue'>humedad óptima ($\theta^*$)</font> permitiría mejorar la simulación de cómo las diferentes coberturas se comportan en situaciones de escasez de agua, de manera que algunas coberturas dejaran de evapotranspirar antes que otras conforme el suelo se seca. Hasta ahora no he utilizado nunca el punto óptimo. Si bien es una mejora en el realismo del funcionamiento de la vegetación, no creo que tenga un gran impacto sobre el balance hídrico de la cuenca.

### 3.2 Evolución de la cobertura del suelo

Primero se calcula el porcentaje de área en el que MODIS define un cambio en los usos del suelo.

```python
# array con las celdas en las que cambia la cobertura del suelo
change = np.zeros(LCmodis.data.shape[1:]) * False
for i in range(LCmodis.data.shape[1]):
    for j in range(LCmodis.data.shape[2]):
        if all(np.isnan(LCmodis.data[:,i,j])):
            change[i,j] = np.nan
        elif len(np.unique(LCmodis.data[:,i,j])) > 1:
            change[i,j] = 1
            
percent_change = np.nansum(change) / np.sum(~np.isnan(change))
print('Área con cambio = {0:.2f} %'.format(percent_change * 100))

# mapa de las celdas con cambio en los usos del suelo
plt.imshow(change)
plt.axis('off');
```

Un 21% de las celdas sufren algún cambio en su clasificación durante los 19 años de datos de MODIS. En el siguiente gráfico se muestran los cambios acontecidos en esas celdas para cada tipo de vegetación.

Los principales cambios que se observan son:

* EL bosque de coníferas a bosque caducifolio y pasto.
* El bosque caducifolio a pasto y, en menor medida, a otros tipos de bosque.
* El pasto a bosque caducifolio.
* Suelo desnudo a pasto.

```python
# evolución del uso del suelo en las celdas con cambio
clases = np.array(list(Amodis.keys())).astype(int)
nrows = len(clases)

fig, axes = plt.subplots(nrows=nrows, figsize=(12, 3.5 * nrows), sharex=True)
claseColor = {i: color for i, color in zip(list(Amodis.keys()), vegColors2)}
for i in range(LCmodis.data.shape[1]):
    for j in range(LCmodis.data.shape[2]):
        if change[i,j] == 1:
            clase = round(LCmodis.data[0,i,j])
            #c = claseColor[clase]
            r = np.where(clase == clases)[0][0]
            axes[r].plot(LCmodis.times, LCmodis.data[:,i,j], color='k', lw=.5, alpha=.25)

for ax, clase in zip(axes, clases):
    ax.set_xticks(LCmodis.times)
    ax.set_xticklabels([time.year for time in LCmodis.times], rotation=90)
    ax.set_yticks(clases)
    ax.set_yticklabels([clasesMODIS[c] for c in clases])
    ax.set_title(clasesMODIS[clase], fontweight='bold')
    
plt.savefig(rutaOut + 'Cambios en los usos del suelo - MODIS.png', dpi=300)
```

#### Series de ET mensual
Calculo la serie mensual de ET para evitar que la variabilidad de la serie 8 d impida ver la periodicidad de la ET. Además, los factores de cultivo se han de calcular mensualmente.

```python
# generar series mensuales
ETsm = serieMensual(ETmodis, agg='sum')
```

De la matriz con las series mensuales de ET extraigo iterativamente las celdas de cada clase de cobertura de suelo cuya clasificación no ha variado en los 19 años de datos de MODIS.

```python
# año
year = 2

ETsm_c = {}
Amodis_ = {}
for lc in Amodis:
    maskLC = LCmodis.data[year,:,:] == lc
    maskLC = np.all(np.stack((change, maskLC), axis=0), axis=0)
    mask3D = np.broadcast_to(maskLC[:-1,:-1], ETsm.data.shape)
    Amodis_[lc] = np.sum(maskLC)
    print(clasesMODIS[lc], Amodis_[lc])
    
    data = ETsm.data.copy()
    data[~mask3D] = np.nan

    ETsm_c[lc] = raster3D(data, ETsm.X, ETsm.Y, ETsm.times, crs=ETsm.crs)
```

Con las series de cada clase creo un gráfico con la serie mensual de ET para cada clase de cobertura. Cada línea del gráfico representa una celda de la cuenca.

```python
r = 50
lw = .25
ylabel = 'ET (mm/mes)'
alpha = 0.25

fig, axes = plt.subplots(nrows=len(ETsm_c), figsize=(17, 3.5 * len(ETsm_c)), sharex=True)

xlim = [min([min(ETsm_c[lc].times) for lc in ETsm_c]),
        max([max(ETsm_c[lc].times) for lc in ETsm_c])]
ymin = 0
ymax = np.ceil(np.nanmax([np.nanmax(ETsm_c[lc].data) for lc in ETsm_c]) / r) * r

for c, (ax, lc) in enumerate(zip(axes, ETsm_c)):
    timex, datax = ETsm_c[lc].times, ETsm_c[lc].data
    for i in range(datax.shape[1]):
        for j in range(datax.shape[2]):
            if np.all(np.isnan(datax[:,i,j])): # celda vacía
                continue
            else:
                ax.plot(timex, datax[:,i,j], lw=lw, c='yellowgreen', alpha=alpha)

    # serie media areal
    ax.plot(timex, np.nanmean(datax, axis=(1, 2)), c='darkolivegreen', lw=4*lw)
    # nº de celdas
    ax.text(0.985, 0.9, 'nº celdas: {0}'.format(Amodis_[lc]), fontsize=12, horizontalalignment='right',
            transform=ax.transAxes)
    # configuración
    ax.tick_params(labelsize=11)
    ax.set(xlim=xlim, ylim=(ymin, ymax))
    ax.set_ylabel(ylabel, fontsize=13)
    ax.set_title(clasesMODIS[lc], fontsize=13, fontweight='bold');
    
plt.savefig(rutaOut + 'ET serie mensual por cobertura del suelo.png', dpi=300)
```

```python
year = 2

ETmm_c = {}
for lc in Amodis:
    maskLC = LCmodis.data[year,:,:] == lc
    maskLC = np.all(np.stack((change, maskLC), axis=0), axis=0)
    mask3D = np.broadcast_to(maskLC[:-1,:-1], ETmm.data.shape)
    
    data = ETmm.data.copy()
    data[~mask3D] = np.nan

    ETmm_c[lc] = raster3D(data, ETmm.X, ETmm.Y, ETmm.times, crs=ETmm.crs)
```

Con las series de cada clase creo un gráfico con la media mensual de ET para cada clase de cobertura. Cada línea del gráfico representa una celda de la cuenca.

```python
ncols = 2
nrows = int(np.ceil(len(ETmm_c)/ncols))
fig, ax = plt.subplots(ncols=ncols, nrows=nrows, figsize=(17, 3.5 * nrows), sharex=True)

xlim = [min([min(ETmm_c[lc].times) for lc in ETmm_c]),
        max([max(ETmm_c[lc].times) for lc in ETmm_c])]
ymin = 0
ymax = np.ceil(np.nanmax([np.nanmax(ETmm_c[lc].data) for lc in ETmm_c]) / r) * r

for i, lc in enumerate(ETmm_c):
    f, c = int(i / ncols), i % ncols
    timex, datax = ETmm_c[lc].times, ETmm_c[lc].data
    for i in range(datax.shape[1]):
        for j in range(datax.shape[2]):
            if np.all(np.isnan(datax[:,i,j])): # celda vacía
                continue
            else:
                ax[f,c].plot(timex, datax[:,i,j], lw=lw, c='yellowgreen', alpha=alpha)

    # serie media areal
    ax[f,c].plot(timex, np.nanmean(datax, axis=(1, 2)), c='darkolivegreen', lw=4*lw)
    # configuración
    ax[f,c].text(0.985, 0.9, 'nº celdas: {0}'.format(Amodis_[lc]), fontsize=12, horizontalalignment='right',
            transform=ax[f,c].transAxes)
    ax[f,c].tick_params(labelsize=11)
    ax[f,c].set(xlim=xlim, ylim=(ymin, ymax))
    ax[f,c].set_ylabel(ylabel, fontsize=13)
    ax[f,c].set_title(clasesMODIS[lc], fontsize=13, fontweight='bold');
    
ax[3,1].axis('off');

plt.savefig(rutaOut + 'ET media mensual por cobertura del suelo.png', dpi=300)
```

__Tabla de factor de cultivo__

Creo una tabla de factores de cultivo a partir de la curva media para cada clase de cobertura del suelo (línea oscura en el gráfico anterior).

Para convertir los datos de ET en factores de cultivo, divido todos los valores de la tabla por el máximo. De esta manera, el máximo valor de factor de cultivo es 1.

```python
# tabla con la ET media mensual de cada clase de cobertura del suelo
FCmodis = pd.DataFrame(index=ETmm_c.keys(), columns=months, dtype=float)
for lc in ETmm_c:
    FCmodis.loc[lc,:] = np.nanmean(ETmm_c[lc].data, axis=(1,2))

FCmodis.index = [clasesMODIS[lc] for lc in FCmodis.index]

# normalizar con el máximo de la tabla
FCmodis /= np.nanmax(FCmodis)
    
FCmodis
```

### 4.3 Comparativa MODIS vs TETIS

Comparo para cada cobertura de suelo la curva de factor de cultivo utilizada en TETIS y la extraida de los datos de ET de MODIS. Sólo puedo hacerlo con las 5 clases de cobertura de suelo coincidentes en ambas fuentes.

```python
claseMODIS = ['deciduous broadleaf trees', 'evergeen needleleaf trees', 'shrub', 'grass', 
              'barren or sparse vegetation']
claseTETIS = ['BLF', 'CNF', 'SSH', 'PAST', 'DEN']

fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(17, 10.5), sharey=True)

for i, (cM, cT) in enumerate(zip(claseMODIS, claseTETIS)):
    f, c = int(i / 2), i % 2
    ax[f,c].plot(FCmodis.loc[cM,:], c='steelblue', label='MODIS')
    ax[f,c].plot(FCtetis.loc[cT,months], c='olive', label='TETIS')
    ax[f,c].set_title(cM, fontsize=13)
    if c == 0:
        ax[f,c].set(ylabel='ET (-)', ylim=(-.05, 1.05))
    
# leyenda
ax[2,1].axis('off');
ax[2,1].legend(*ax[f,c].get_legend_handles_labels(), loc=4);

plt.savefig(rutaOut + 'Factor de culvigo. MODIS vs TETIS.png', dpi=300)
```

Las curvas del factor de cultivo utilizadas en TETIS y las extraídas de MODIS tienen poco que ver. Las mayores diferencias están en los árboles perennes (que no tienen un factor constante), el matorral (que tiene un factor de cultivo inferior) y el suelo desnudo (que no tiene un factor de cultivo nulo durante el verano y otoño).

Para completar la tabla del factor de cultivo que requiere TETIS faltan dos clases: suelo antrópico y cuerpos de agua. Los cuerpos de agua deberían tener un factor de cultivo de 1, puesto que la evaporación podrá ser siempre la potencial. El suelo antrópico no está tan claro; desde luego no puede ser el 0 he utilizado hasta ahora; como aproximación se puede utilizar la misma curva que tiene el suelo desnudo.
