## 2 Convertir ASCII en netCDF

```python
# ruta del proyecto TETIS
rutaTETIS = 'F:/Proyectos/GESDIVAH/SDM/TETIS/PdE/Actual/Bias/'
rutaTETIS = 'C:/Users/jcr693/OneDrive - UNICAN - Estudiantes/Proyectos/GESDIVAH/SDM/TETIS/PdE/Actual/Bias/'

# lista de archivos con los resultados
os.chdir(rutaTETIS + '_ASCII/')
files = [file for file in os.listdir() if file.endswith('asc') & file.startswith(varTETIS)]
```

```python
# cargar archivos ascii con los resultados de TETIS
for i, file in enumerate(files):
    print('archivo {0:>3} de {1}: {2}'.format(i+1, len(files), file), end='\r')
    # leer archivo ascii
    asc = read_ascii(file, crs=25830)
    if i == 0:
        Datatetis = asc.data
        atrsTetis = asc.attributes
    else:
        Datatetis = np.dstack((Datatetis, asc.data))    

# transformar 'Datatetis' para que la primera dimensión sea el tiempo
Datatetis = np.moveaxis(Datatetis, 2, 0)
```

```python
# serie de fechas
times = [datetime.strptime(f[3:11], '%Y%m%d').date() for f in files]
```

```python
# guardar datos como raster3D
TETIS = raster3D(Datatetis, asc.X, asc.Y, times, units='mm',
                 variable='modelled evapotranspiration', label='ET',
                 crs=CRS.from_epsg(25830))
```

```python
# fijar rutas y nombre del archivo
rutaData = rutaBase + '../data/'
filename = 'TETIS_' + var + '_100m.nc'

# exportar los datos como netCDF
description = 'Serie temporal de mapas de evapotranspiración simulados con TETIS para la cuenca del Deva'
MODIS2netCDF(rutaData + filename, TETIS, description)
```
